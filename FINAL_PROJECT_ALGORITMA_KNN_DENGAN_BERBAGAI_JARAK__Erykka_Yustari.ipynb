#**Model K-Nearest Neighbor (KNN)**

Model K-Nearest Neighbor (KNN) adalah model Machine Learning yang dapat melakukan klasifikasi sebuah data didasarkan pada sifat-sifat data didekatnya. Sebuah ungkapan yang bisa membantu dengan mudah memahami ide dasar algoritma ini adalah: "Jika kamu ingin mengetahui sifat seseorang, lihatnya sifat-sifat orang terdekatnya. Sifatnya tidak akan jauh berbeda dari mereka."

Gambar di bawah ini dapat meilustrasikan hal di atas dengan jelas

Screenshot (4).png


Data baru yang dilambangkan dengan simbol segitiga biru dapat diklasifikasikan sebagai kelompok kotak hijau atau bintang merah jambu. Jika dilihat berdasarkan kriteria jumlah k=1 maka data segitiga biru diklasifikasikan ke dalam kelompok kotak hijau. Sedangkan jika dilihat berdasarkan kriteria jumlah k=3 maka data tersebut diklasifikasikan sebagai kelompok bintang merah jambu. Hal yang sama juga dapat diterapkan untuk k=9. Sebenarnya, kita telah melakukan klasifikasi berdasarkan jarak. Kedekatan yang dimaksud ketika k=1, k=3, dan k=9, kita selalu mengukur kedekatan antara data baru(segitiga biru) dengan data-data berlabel kelompok sebelumnya. Berdasarkan ide dasar di atas, maka diturunkanlah dalam sebuah algoritma untuk menentukan kelompok data dari sebuah inputan data baru(segitiga biru).

# **FUNGSI JARAK**

Perhitungan jarak banyak digunakan dalam menentukan tingkat kemiripan atau tidaknya dua buah vektor. Sehingga metode ini banyak digunakan untuk melakukan pengenalan pola (Wurdianarto et al, 2014). Perbedaan perhitungan jarak terdekat/ jarak kemiripan sangat tepat digunakan untuk menganalisis kelas perbedaan. Perhitungan jarak terdekat/jarak kemiripan menggunakan beberapa nilai matriks biasanya digunakan untuk mengekstrak kemiripan objek data dan dibantu dengan proses klasifikasi menggunakan algoritma yang efiesien. Untuk dataset klasifikasi, beberapa metode perhitungan jarak terdekat/jarak kemiripan adalah Manhattan, Minkowski, Chebishev, Bray-Curtiz, Canbera.

#**Evaluasi Model Machine learning**

Mengukur Kinerja Algoritma Klasifikasi dengan Confusion Matrix
* Pengukuran terhadap kinerja suatu sistem klasifikasi
merupakan hal yang penting. Kinerja sistem klasifikasi
menggambarkan seberapa baik sistem dalam
mengklasifikasikan data.
* Confusion matrix merupakan salah satu metode yang
dapat digunakan untuk mengukur kinerja suatu metode
klasifikasi. 
* Pada dasarnya confusion matrix mengandung informasi
yang membandingkan hasil klasifikasi yang dilakukan
oleh sistem dengan hasil klasifikasi yang seharusnya 
* Pengukuran kinerja menggunakan confusion matrix,
terdapat 4 (empat) istilah :

1.   True Positive (TP)
2.   True Negative (TN)
3. False Positive (FP) 
4. False Negative (FN)

TP yaitu jumlah data positif yang terklasifikasi dengan benar oleh sistem.
TN yaitu jumlah data negatif yang terklasifikasi dengan benar oleh sistem.
FN yaitu jumlah data negatif namun terklasifikasi salah oleh sistem.
FP yaitu jumlah data positif namun terklasifikasi salah oleh sistem.

##**Tabel confusion matrix**
* Precision adalah tingkat ketepatan antara informasi yang diminta oleh pengguna dengan jawaban yang diberikan oleh sistem. 
* Recall adalah tingkat keberhasilan sistem dalam menemukan kembali sebuah informasi.

**Contoh Penerapan Confusion Matrix**


```python
import pandas as pd

data = {'y_Actual':    [1, 0, 0, 1, 0, 1, 0, 0, 1, 0, 1, 0],
        'y_Predicted': [1, 1, 0, 1, 0, 1, 1, 0, 1, 0, 0, 0]
        }

df = pd.DataFrame(data, columns=['y_Actual','y_Predicted'])
print (df)
```

```python
confusion_matrix = pd.crosstab(df['y_Actual'], df['y_Predicted'], rownames=['Actual'], colnames=['Predicted'])
print (confusion_matrix)
```

**Menampilkan confusion matrix menggunakan seaborn**

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

data = {'y_Actual':    [1, 0, 0, 1, 0, 1, 0, 0, 1, 0, 1, 0],
        'y_Predicted': [1, 1, 0, 1, 0, 1, 1, 0, 1, 0, 0, 0]
        }

df = pd.DataFrame(data, columns=['y_Actual','y_Predicted'])
confusion_matrix = pd.crosstab(df['y_Actual'], df['y_Predicted'], rownames=['Actual'], colnames=['Predicted'])

sn.heatmap(confusion_matrix, annot=True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Untuk memahami konsep, kita akan menggunakan data X dan Y seperti yang kita tunjukkan pada proses penghitungan manual

```python
X=np.array([[1,1], [2,2], [3,2], [12,11], [11,13], [14,15]])
Y=np.array([0,0,0,1,1,1])
x_baru=np.array([9,10])
```

```python
X
```

```python
Y
```

```python
x_baru
```

**Langkah 1:**

Menentukan parameter k (jumlah tetangga paling dekat)

```python
k=3
```

**Langkah 2:**

Menghitung kuadrat jarak eucliden objek terhadap data training yang diberikan.

```python
def jarak_euclidean(x,y):
    return np.sqrt(sum(pow(a-b, 2) for a, b in zip(x,y)))
```

```python
jarak=[]
for x_i in X:
    jarak_euclid=jarak_euclidean(x_i, x_baru)
    jarak.append(jarak_euclid)
    print(jarak)
print(jarak)
```

**Langkah 3:**

Mengurutkan hasil langkah 2 secara ascending (berurutan dari rendah ke tinggi)

```python
urut=np.sort(jarak)
list_urut=[]
for i in range(len(urut)):
    list_urut.append(urut[i])
print(list_urut)
```

```python
urutan=[]
for i in list_urut:
    urtn=jarak.index(i)
    urutan.append(urtn)
print(urutan)
```

atau dengan sederhana dapat menggunakan *np.argsort* seperti berikut:

```python
np.argsort(jarak)
```

**Langkah 4:**

Mengumpulkan kategori data (Klasifikasi nearest neightbor berdasarkan nilai k)

Kelas data 3 data-data terdekat (k=3)

```python
ambil=urutan[:k]
print(Y[ambil])
```

**Langkah 5:**

Dengan menggunakan kategori nearest neighbor yang paling mayoritas maka dapat diprediksikan kategori objek

```python
from collections import Counter
kelas_paling_banyak = Counter(Y[ambil]).most_common(1)
kelas_paling_banyak
```

## **KNN DENGAN JARAK MINKOWSKI**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=np.array([[1,1], [2,2], [3,2], [12,11], [11,13], [14,15]])
Y=np.array([0,0,0,1,1,1])
x_baru=np.array([9,10])
```

```python
X
```

```python
Y
```

```python
x_baru
```

**Langkah 1:**

Menentukan parameter k (jumlah tetangga paling dekat)

```python
k=3
```

**Langkah 2:**

Menghitung kuadrat jarak minkowski objek terhadap data training yang diberikan.

```python
def akar_ke_n(nilai, akar_n):
  nilai_akar=1/float(akar_n)
  return round(float(nilai)**float(nilai_akar),3)

def minkowski(x,y,p_value):
  return akar_ke_n(sum(pow(abs(a-b), p_value) for a, b in zip(x,y)), p_value)
```

```python
jarak=[]
for x_i in X:
    jarak_minkowski=minkowski(x_i, x_baru,1)
    jarak.append(jarak_minkowski)
    print(jarak)
print(jarak)
```

**Langkah 3:**

Mengurutkan hasil langkah 2 secara ascending (berurutan dari nilai tinggi ke rendah)

```python
urut=np.sort(jarak)
list_urut=[]
for i in range(len(urut)):
    list_urut.append(urut[i])
print(list_urut)
```

```python
urutan=[]
for i in list_urut:
    urtn=jarak.index(i)
    urutan.append(urtn)
print(urutan)
```

atau dengan sederhana dapat menggunakan *np.argsort* seperti berikut:

```python
np.argsort(jarak)
```

**Langkah 4:**

Mengumpulkan kategori data (Klasifikasi nearest neightbor berdasarkan nilai k)

Kelas data 3 data data terdekat (k=3)

```python
ambil=urutan[:k]
print(Y[ambil])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=np.array([[1,1], [2,2], [3,2], [12,11], [11,13], [14,15]])
Y=np.array([0,0,0,1,1,1])
x_baru=np.array([9,10])
```

```python
X
```

```python
k=3
```

**Langkah 2:**

Menghitung kuadrat jarak minkowski objek terhadap data training yang diberikan.

```python
def chebyshev_distance(a, b):
    return max(abs(e1-e2) for e1, e2 in zip(a,b))
```

```python
jarak=[]
for x_i in X:
    jarak_chebyshev_distance=chebyshev_distance(x_i, x_baru)
    jarak.append(jarak_chebyshev_distance)
    print(jarak)
print(jarak)
```

Menuliskan algoritma KNN dalam bentuk fungsi

```python
def manhattan(x,y): # Perhitungan Jarak Manhattan
  return sum(abs(a-b) for a,b in zip(x,y))

def akar_ke_n(nilai, akar_n): # Perhitungan Jarak Minkowski
  nilai_akar=1/float(akar_n)
  return round(float(nilai)**float(nilai_akar),3)

def minkowski(x,y,p_value):
  return akar_ke_n(sum(pow(abs(a-b), p_value) for a, b in zip(x,y)), p_value)

def chebyshev_distance(a, b): # Perhitungan Jarak Chebyshev
  return max(abs(e1-e2) for e1, e2 in zip(a,b))

def Bray_Curtiz(x,y): # Perhitungan Jarak Bray_Curtiz
  return sum(abs(a-b)/(a+b) for a,b in zip(x,y))

def canbera(x,y): # Perhitungan Jarak Canbera
  return sum(abs(a-b)/(abs(a)+abs(b)) for a, b in zip(x,y))
```

```python
def _knn(x):
    # Menghitung jarak antara x baru dengan semua sampel di data latih
    jarak = [manhattan(x, x_latih) for x_latih in X_latih]
    
    #Mengurutkan data dan mengambil sejumlah k objek
    k_idx=np.argsort(jarak)[:k]
    
    #Mengambil kelas dari k tetangga terdekat dari data latih
    k_neighbor_labels = [y_latih[i] for i in k_idx]
    
    #Mengambil kelas yang paling sering muncul
    kelas_paling_banyak = Counter(k_neighbor_labels).most_common(1)
    return kelas_paling_banyak[0][0]

def KNN(X_latih, y_latih, k, x_baru):
    y_pred=[_knn(x_i) for x_i in x_baru]
    return np.array(y_pred)
```

```python
def hitung_tp_tn_fn_fp(y_asli, y_pred):
  tp = sum((y_asli == 1)&(y_pred ==1))
  tn = sum((y_asli == 0)&(y_pred ==0))
  fn = sum((y_asli == 1)&(y_pred ==0))
  fp = sum((y_asli == 0)&(y_pred ==1))
  return tp, tn, fp, fn
```
